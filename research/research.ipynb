```python
## 2 Now the explanation for ensure_notation

def mul(x:int,y:int) -> int:
    return x*y
```

```python
mul(2,2)
```

```python
mul(2,"4")
```

```python
from ensure import ensure_annotations

@ensure_annotations
def mul(x:int,y:int) -> int:
    return x*y
```

```python
mul(2,"4")
```
